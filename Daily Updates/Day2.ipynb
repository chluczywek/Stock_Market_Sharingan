```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
```

```python
df_d=pd.read_csv('drive/MyDrive/Stock_Market_Sharingan/jpm_us_d.csv')
df_w=pd.read_csv('drive/MyDrive/Stock_Market_Sharingan/jpm_us_w.csv')
df_m=pd.read_csv('drive/MyDrive/Stock_Market_Sharingan/jpm_us_m.csv')
df_q=pd.read_csv('drive/MyDrive/Stock_Market_Sharingan/jpm_us_q.csv')
df_y=pd.read_csv('drive/MyDrive/Stock_Market_Sharingan/jpm_us_y.csv')

df_d.head()
```

```python
df_d['Datetime_d'] = pd.to_datetime(df_d['Date_d'])
df_m['Datetime_m'] = pd.to_datetime(df_m['Date_m'])
df_y['Datetime_y'] = pd.to_datetime(df_y['Date_y'])
```

```python
# Extract year
df_d['year'] = df_d['Datetime_d'].dt.year
df_m['year'] = df_m['Datetime_m'].dt.year
df_y['year'] = df_y['Datetime_y'].dt.year

# Extract month
df_d['month'] = df_d['Datetime_d'].dt.month
df_m['month'] = df_m['Datetime_m'].dt.month
df_y['month'] = df_y['Datetime_y'].dt.month

# Perform the join
merged_df = pd.merge(df_d, df_m, on=['year', 'month'], how='inner')
merged_df = pd.merge(merged_df, df_y, on='year', how='inner')
merged_df.head()
```

```python
# Get column names as a list
column_list = merged_df.columns.tolist()

# Print the list of columns
print(column_list)
```
